```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
# Create input universe variables
x = np.array([0, 1, 2, 3])
# Define membership degrees for fuzzy set A manually
A = [0.2,0.3,0.4,0.5]

# Define membership degrees for fuzzy set B manually
B = [0.1,0.2,0.2,1]
```

```python
union_result = np.fmax(A, B)
intersection_result = np.fmin(A, B)
complement_A = 1 - np.array(A)
complement_B = 1 - np.array(B)
```

```python
union_result
```

```python
intersection_result
```

```python
complement_A
```

```python
complement_B
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, A, label='A')
plt.plot(x, B, label='B')
plt.plot(x, union_result, label='Union (A OR B)')
plt.plot(x, intersection_result, label='Intersection (A AND B)')
plt.plot(x, complement_A, label='Complement of A')
plt.plot(x, complement_B, label='Complement of B')

# Add labels and legend
plt.xlabel('Input Universe')
plt.ylabel('Membership Degree')
plt.title('Fuzzy Set Operations')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
algebraic_sum = [A + B - A * B for A, B in zip(A, B)]
algebraic_difference = [A * B for A, B in zip(A, B)]
bounded_sum = [min(A + B, 1) for A, B in zip(A, B)]
bounded_difference = [max(A - B, 0) for A, B in zip(A, B)]
```

```python
algebraic_sum
```

```python
algebraic_difference
```

```python
bounded_sum
```

```python
bounded_difference
```

```python
x = np.arange(0, len(A))
plt.figure(figsize=(12, 8))
plt.plot(x, algebraic_sum, label='Algebraic Sum (A + B - A * B)')
plt.plot(x, algebraic_difference, label='Algebraic Difference (A * B)')
plt.plot(x, bounded_sum, label='Bounded Sum (min(A + B, 1))')
plt.plot(x, bounded_difference, label='Bounded Difference (max(A - B, 0))')

# Add labels and legend
plt.xlabel('Input Universe Index')
plt.ylabel('Result')
plt.title('Fuzzy Set Operations')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
