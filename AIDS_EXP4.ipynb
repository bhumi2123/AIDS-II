```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Loan_Status', data=df)
plt.title('Distribution of Loan Status')
plt.show()
```

```python

sns.countplot(x='Property_Area', data=df)
plt.title('Distribution of Property Area ')
plt.show()
```

```python
sns.boxplot(y='ApplicantIncome', data=df)
plt.title('Distribution of Applicant Income')
plt.show()
```

```python
sns.histplot(df['LoanAmount'], kde=True)
plt.title('Distribution of Loan Amount')
plt.show()
```

```python
sns.countplot(x='Credit_History', hue='Loan_Status', data=df)
plt.title('Credit History vs. Loan Status')
plt.show()
```

```python
sns.countplot(x='Property_Area', hue='Loan_Status', data=df)
plt.title('Property Area vs. Loan Status')
plt.show()
```

```python
sns.boxplot(x='Education', y='LoanAmount', data=df)
plt.title('Education vs. Loan Amount')
plt.show()
```

```python
corr = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
```
